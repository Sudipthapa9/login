```python
num = 2

while num <= 100:
    i = 2
    is_prime = True

    while i <= num // 2:
        if num % i == 0:
            is_prime = False
            break
        i += 1

    if is_prime:
        print(num)

    num += 1
```

```python
#print prime numbers between 1 and 100 using while loop --- IGNORE ---
for num in range (2, 101):
    is_prime = True
    for i in range(2, int(num **0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end="")
```

```python
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")

print("Loop finished.")
```

```python
numbers = [1, 2, 3, 4, 5]

i = 0
while i < len(numbers) and not (numbers[i] % 2 == 0):
    i += 1

if i < len(numbers):
    print(f"{numbers[i]} is even")
else:
    print("No even numbers found.")
```
