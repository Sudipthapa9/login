```python
i = 1
while i <= 10:
    print(f"5 x {i} = {5*i}")
    i += 1
```

```python
num = 2

while num <= 100:
    i = 2
    is_prime = True

    while i <= num // 2:
        if num % i == 0:
            is_prime = False
            break
        i += 1

    if is_prime:
        print(num)

    num += 1
```

```python
#print prime numbers between 1 and 100 using while loop --- IGNORE ---
for num in range (2, 101):
    is_prime = True
    for i in range(2, int(num **0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end="")
```
