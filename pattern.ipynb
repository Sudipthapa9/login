```python
for i in range(5, 0, -1):
    for j in range(5, 5 - i, -1):
        print(j, end="")
    print()
```

```python
for i in range(1, 6):
    for j in range(5, 5 - i, -1):
        print(j, end=" ")
    print(" ")
```

```python
for i in range(1, 6):
    for j in range(6 - i):
        print(i, end="")
    print()
```
